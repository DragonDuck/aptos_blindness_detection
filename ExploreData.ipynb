This notebook explores the data and its attributes

```python
import os
from PIL import Image
import numpy as np
import pandas as pd
```

```python
img_dir = "train_images/"
```

```python
file_list = [img_dir + f for f in os.listdir(img_dir)]
print("There are {} images.".format(len(file_list)))
```

```python
img_sizes = []
for file in file_list:
    img = Image.open(file)
    img_sizes.append((img.width, img.height))
print("Unique image dimensions:")
for entry in set(img_sizes):
    print(" - " + str(entry))
```
